```python
numbers = [34, 10, 4, 6, 10, 23, 90, 100, 21, 35, 95, 1, 36, 38, 19, 1, 6, 87, 1000, 13456, 360]
result = {x : [j for j in range(1, x +1) if x % j == 0] for x in numbers }
print(result)
```

```python
numbers = [34, 10, 4, 6, 10, 23, 90, 100, 21, 35, 95, 1, 36, 38, 19, 1, 6, 87, 1000, 13456, 360]
result = {}
for i in numbers:
    j = [j for j in range(1,i +1) if i % j == 0]
    result[i] = result.get(i, j)
print(result) 
```
